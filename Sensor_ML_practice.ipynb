```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path(r"C:\Users\demki\Desktop\Python\Data\Git files/UCI HAR Dataset")

# Load feature names
features = pd.read_csv(
    DATA_PATH / "features.txt",
    sep="\s+",
    header=None,
    names=["index", "feature"]
)

feature_names = features["feature"].values

# Load activity labels
activity_labels = pd.read_csv(
    DATA_PATH / "activity_labels.txt",
    sep="\s+",
    header=None,
    names=["id", "activity"]
)

# Load TRAIN
X_train = pd.read_csv(DATA_PATH / "train/X_train.txt", sep="\s+", header=None)
y_train = pd.read_csv(DATA_PATH / "train/y_train.txt", sep="\s+", header=None)
subject_train = pd.read_csv(DATA_PATH / "train/subject_train.txt", sep="\s+", header=None)

# Load TEST
X_test = pd.read_csv(DATA_PATH / "test/X_test.txt", sep="\s+", header=None)
y_test = pd.read_csv(DATA_PATH / "test/y_test.txt", sep="\s+", header=None)
subject_test = pd.read_csv(DATA_PATH / "test/subject_test.txt", sep="\s+", header=None)

# Assign column names
X_train.columns = feature_names
X_test.columns = feature_names

# Merge datasets
X = pd.concat([X_train, X_test])
y = pd.concat([y_train, y_test])
subjects = pd.concat([subject_train, subject_test])

# Combine into one dataframe
df = X.copy()
df["activity_id"] = y
df["subject"] = subjects

# Add activity names
df = df.merge(activity_labels, left_on="activity_id", right_on="id", how="left")
df.drop(columns=["id"], inplace=True)

df.head()
```
